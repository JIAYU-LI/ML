```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
PATH = os.getcwd()
CAT_DATA_PATH = os.path.join(PATH, 'large_categorical_data.csv')
categorical_data = pd.read_csv(CAT_DATA_PATH, delimiter=',')
print(categorical_data.shape)
```

```python
del categorical_data['Unnamed: 0']
```

```python
THRESHOLD = 100

for i in range(14755, 15000):
    var_name = 'Var'+str(i)
    try:
        variances = categorical_data.loc[:,var_name]  
    except KeyError:
        pass
    else:
        uni_len = len(variances.unique())
        if uni_len<2 or uni_len>THRESHOLD:
            del categorical_data[var_name]
    
print(categorical_data)
```

```python
categorical_data_onehot = pd.get_dummies(categorical_data)
print(categorical_data_onehot.shape)
```

```python
categorical_data_onehot.to_csv("large_cate_onehot.csv")
```
