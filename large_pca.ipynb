```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
def readin_large_data():
    """Read small data set into a 2D numpy array
    return: numpy array contains the small data set, shape: (50,000, 230)
    """
    PATH = os.getcwd()
    
    DATA_PATH = os.path.join(PATH, 'large_clean_numerical_data.csv')

    data = pd.read_csv(DATA_PATH, delimiter = ',')
    
    del data["Unnamed: 0"]
    
    return data

# data_list = []

# for i in range(1,6):
#     if i == 1:
#         data = readin_large_data(i)
#     else:
#         data = pd.concat((data, readin_large_data(i)), ignore_index=True)

# data = pd.concat(data_list)

data = readin_large_data()

print(data)
```

```python
from sklearn.decomposition import PCA

data_pca = PCA(n_components=1000).fit(data.T)
print(data_pca.components_.T.shape)
```

```python
print(data_pca.components_.T)
```

```python
large_num_pca = pd.DataFrame(data_pca.components_.T)
print(large_num_pca)
```

```python
large_num_pca.to_csv('large_numerical_pca.csv')
```

```python
large_cat = pd.read_csv('large_cate_onehot.csv')
```

```python
del large_cat["Unnamed: 0"]
```

```python
large_data_clean = np.hstack((data_pca.components_.T,large_cat))
print(large_data_clean.shape)
```
